```python
import os
import re
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from sklearn.model_selection import train_test_split
from pyvi import ViTokenizer
```

```python
def load_stopwords(filepath):
    with open(filepath, 'r', encoding='utf-8') as f:
        stop_words = set(f.read().splitlines())
    return stop_words

# Hàm đọc một số lượng file txt nhất định trong thư mục
def load_text_files(folder_path, num_files=2):
    texts = []
    files = os.listdir(folder_path)[:num_files]  # Lấy số lượng file nhất định
    for filename in files:
        if filename.endswith(".txt"):
            with open(os.path.join(folder_path, filename), 'r', encoding='utf-8') as file:
                texts.append(file.read())
    return texts

# Tiền xử lý dữ liệu văn bản
def preprocess_text(text, stop_words):
    text = text.lower()
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^\w\s]', '', text)  # Loại bỏ ký tự không phải chữ cái/số
    text = re.sub(r'(\d+)', '<number>', text)  # Thay số bằng token <number>
    text = ViTokenizer.tokenize(text)  # Tách từ tiếng Việt
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

# Load và tiền xử lý toàn bộ dữ liệu
dataset_folder = "/kaggle/input/10000-vietnamese-books/output"
stop_words = load_stopwords('/kaggle/input/vietnamesestopwords-txt/vietnamese-stopwords.txt')


texts = load_text_files(dataset_folder, num_files=100)

# Tiền xử lý tất cả các văn bản
processed_texts = [preprocess_text(text, stop_words) for text in texts]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

from keras.preprocessing.sequence import pad_sequences

def tokenize_texts(texts, max_sequence_len=None):
    tokenizer = Tokenizer(oov_token="<OOV>")
    tokenizer.fit_on_texts(texts)
    sequences = tokenizer.texts_to_sequences(texts)

    if max_sequence_len is None:
        max_sequence_len = min(500, max([len(seq) for seq in sequences]))  # Giới hạn độ dài chuỗi tối đa

    sequences = pad_sequences(sequences, maxlen=max_sequence_len, padding='post')
    return sequences, tokenizer, max_sequence_len

def tokenize_texts_in_batches(texts, batch_size=1000, max_sequence_len=None):
    tokenizer = Tokenizer(oov_token="<OOV>")
    tokenizer.fit_on_texts(texts)

    sequences = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size]
        batch_sequences = tokenizer.texts_to_sequences(batch_texts)
        if max_sequence_len is None:
            max_sequence_len = min(500, max([len(seq) for seq in batch_sequences]))
        batch_sequences = pad_sequences(batch_sequences, maxlen=max_sequence_len, padding='post')
        sequences.extend(batch_sequences)

    return np.array(sequences), tokenizer, max_sequence_len

def tokenize_texts_with_limit(texts, max_words=20000, max_sequence_len=None):
    tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
    tokenizer.fit_on_texts(texts)
    sequences = tokenizer.texts_to_sequences(texts)

    if max_sequence_len is None:
        max_sequence_len = min(500, max([len(seq) for seq in sequences]))  # Giới hạn độ dài chuỗi tối đa

    sequences = pad_sequences(sequences, maxlen=max_sequence_len, padding='post')
    return sequences, tokenizer, max_sequence_len

sequences, tokenizer, max_sequence_len = tokenize_texts_with_limit(texts, max_words=20000, max_sequence_len=500)
```

```python
import pickle
# Lưu lại tokenizer
def save_tokenizer(tokenizer, path='tokenizer_V1.pkl'):
    with open(path, 'wb') as handle:
        pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)

save_tokenizer(tokenizer)
```

```python
# Khởi tạo mô hình
def create_lstm_model(vocab_size, embedding_dim=128, max_sequence_len=500):
    model = Sequential()
    model.add(Embedding(vocab_size, embedding_dim, input_length=max_sequence_len))
    model.add(LSTM(256, return_sequences=True))
    model.add(Dropout(0.2))
    model.add(LSTM(128))
    model.add(Dropout(0.2))
    model.add(Dense(vocab_size, activation='softmax'))
    model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
from tensorflow.keras.layers import Dropout

# Tạo mô hình
vocab_size = len(tokenizer.word_index) + 1  # Tính toán kích thước từ vựng
model = create_lstm_model(vocab_size)
```

```python

# Chuẩn bị dữ liệu cho training
def prepare_sequences(sequences, sequence_length):
    X, y = [], []
    for seq in sequences:
        for i in range(1, len(seq)):
            X.append(seq[:i])
            y.append(seq[i])
    return pad_sequences(X, maxlen=sequence_length, padding='pre'), np.array(y)

X, y = prepare_sequences(sequences, max_sequence_len)
```

```python

# train model
history = model.fit(X, y, epochs=10, batch_size=32, validation_split=0.1)
```

```python
# Hàm generate_text
def generate_text(model, tokenizer, max_sequence_len, seed_text, n_words):
    result = seed_text
    for _ in range(n_words):
        token_list = tokenizer.texts_to_sequences([result])[0]
        token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
        predicted = model.predict(token_list, verbose=0)
        predicted_word_index = np.argmax(predicted, axis=-1)[0]
        predicted_word = tokenizer.index_word[predicted_word_index]
        result += " " + predicted_word
    return result
```
